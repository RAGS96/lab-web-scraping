#### 3. Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.
Hint: If you hover over the title of the movie, you should see the director's name. Can you find where it's stored in the html?

````python
# This is the url you will scrape in this exercise 
url = "https://www.imdb.com/chart/top"
````

````python
response = requests.get(url).content

soup = BeautifulSoup(response, features = "lxml")
soup.prettify()

result = soup.find_all("td", attrs={"class":"titleColumn"})
#display(result)

imbd_top = []
for movies in result:
    movie_name = "".join(re.findall("[A-za-z]+ ?", movies.text)) #? -> not sure if this make a difference and what it does exactly
    initial_release = re.findall("[0-9]{4}", movies.text)[-1]
    director_stars = movies.a["title"] 
    imbd_top.append({"Movie Name":movie_name, "Release":initial_release, "Director & Cast":director_stars})
    
imbd_top_df = pd.DataFrame(imbd_top)
display(imbd_top_df)


#finally...
````
